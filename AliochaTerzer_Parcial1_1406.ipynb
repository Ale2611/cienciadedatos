```python
import pandas as pd
import matplotlib.pyplot as plt


# Leer los datos del archivo Excel
archivo = ("c2022_tp_c_resumen.xlsx")
datoscontodo = pd.read_excel(archivo)

indice_a_eliminar = 0  # Por ejemplo, supongamos que es la fila número 10

# Eliminar la fila del DataFrame
datos = datoscontodo.drop(indice_a_eliminar)
datos.head()

```

```python
#1) Visualizar el archivo cargado (lista). (10 p)
def visualizar():
  display(datos)



```

```python
def top_5_provincias():
  datos["Total Población"] = datos["Total de población de varones"] + datos["Total de población de mujeres"]
  datos_ordenados = datos.sort_values(by="Total Población", ascending=False)
  top_5_provincias = datos_ordenados.head(5)
  top_5_provincias_asc = top_5_provincias.sort_values(by="Total Población", ascending=True)
  display(top_5_provincias_asc[['Jurisdicción', 'Total Población']])
```

```python
def diagra():
  datos["Total Población"] = datos["Total de población de varones"] + datos["Total de población de mujeres"]
  datos_ordenados = datos.sort_values(by="Total Población", ascending=False)
  top_3_provincias = datos_ordenados.head(3)


  plt.figure(figsize=(10, 6))

  plt.bar(top_3_provincias['Jurisdicción'], top_3_provincias['Total Población'], color='skyblue')
  plt.xlabel('Provincia')
  plt.ylabel('Total Población')
  plt.title('Top 3 Provincias con Más Población')
  plt.grid(axis='y', linestyle='--')



  plt.tight_layout()
  plt.show()
```

```python
texto="""Ingrese la operación a realizar"

1 Vizualizar el archivo
2 Visualizar en una lista las 5 provincia con mayor población
3 Visualizar en un solo diagrama de barras las 3 provincias con mayor población.

"""

opcion=int(input(texto))

   if opcion == 1:
        visualizar()
    elif opcion == 2:
        top_5_provincias()
    elif opcion == 3:
        diagra()
    else:
      print("ingresar una opción correcta"):
    opcion = int(input(texto))
```
