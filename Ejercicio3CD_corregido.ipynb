```python
import pandas as pd #importación de la librería panda
import numpy as np
import matplotlib.pyplot as plt
archivo = 'Influencer.xlsx' #Asigna el nombre del archivo Excel a la variable archivo
dataframe = pd.read_excel(archivo, sheet_name='Hoja1') #: Utiliza la función read_excel de pandas para leer el archivo Excel especificado por la variable archivo. La hoja de Excel llamada 'Hoja1' se carga en un DataFrame de pandas, que se almacena en la variable dataframe.

#Renombrar las columnas del DataFrame
nuevos_nombres = ['N.º', 'Usuario', 'Propietario', 'Seguidores_en_millones', 'Ocupación', 'País', 'Cuenta_de_marca']
dataframe.columns = nuevos_nombres


dataframe['Cuenta_de_marca'] = dataframe['Cuenta_de_marca'].replace({'Sí': 1, 'No': 0})
#Se reemplazan los valores en la columna 'Cuenta_de_marca': 'Sí' por 1 y 'No' por 0. Esto es útil para trabajar con datos numéricos en lugar de cadenas de texto.

def mas_seguidores():
    indice_max_seguidores = dataframe['Seguidores_en_millones'].idxmax()
    celebridad_mas_seguidores = dataframe.loc[indice_max_seguidores]

    nombre = celebridad_mas_seguidores['Propietario']
    seguidores = celebridad_mas_seguidores['Seguidores_en_millones']
    pais = celebridad_mas_seguidores['País']

    print("Nombre:", nombre)
    print("Cantidad de seguidores:", seguidores, "millones")
    print("País de origen:", pais)

def paises():
    conteo_paises = dataframe['País'].value_counts()
    top_paises = conteo_paises.head(3)
    print(top_paises)

def marca():
    cuentas_de_marca = dataframe[dataframe['Cuenta_de_marca'] == 1]
    indice_max_seguidores_marca = cuentas_de_marca['Seguidores_en_millones'].idxmax()
    max_seguidores_marca = dataframe.loc[indice_max_seguidores_marca]
    nombre = max_seguidores_marca['Propietario']
    seguidores = max_seguidores_marca['Seguidores_en_millones']
    print("Nombre:", nombre)
    print("Cantidad de seguidores:", seguidores, "millones")


```

```python
def grafi():
     # Agrupa los datos por 'País' y suma los valores de 'Seguidores_en_millones' para cada país
    seguidores_por_pais = dataframe.groupby('País')['Seguidores_en_millones'].sum()
    x = seguidores_por_pais.index     # Obtiene los nombres de los países (índices del DataFrame resultante) y los guarda en la variable x
    y = seguidores_por_pais.values     # Obtiene los valores de los seguidores (valores del DataFrame resultante) y los guarda en la variable y
    plt.figure(figsize=(10, 6))     # Configura el tamaño de la figura del gráfico (ancho: 10 pulgadas, alto: 6 pulgadas)
      # Crea un gráfico de barras con los países en el eje x y el número de seguidores en el eje y
    # color='skyblue' establece el color de las barras
    plt.bar(x, y, color='skyblue')
    plt.xlabel('País')     # Establece la etiqueta del eje x
    plt.ylabel('Seguidores en millones') # Establece la etiqueta del eje y
    plt.title('Total de seguidores por país') # Establece el título del gráfico
    plt.xticks(rotation=45)    # Ajusta el diseño del gráfico para que todos los elementos se acomoden de manera adecuada

    plt.tight_layout()
    plt.show()
    grafi()
```

```python
def grafico_pais():

      # Elimina los caracteres de espacio no rompible y los espacios adicionales
    dataframe['País'] = dataframe['País'].str.replace('\xa0', ' ').str.strip()

     # Normaliza los valores del país
    dataframe['País'] = dataframe['País'].replace({
        'EEUU': 'EstadosUnidos',
        'Estados Unidos': 'EstadosUnidos',
        'USA': 'EstadosUnidos',
        'United States': 'EstadosUnidos',
        'U.S.A.': 'EstadosUnidos',
        'U.S.': 'EstadosUnidos'
    })
  # Cuenta las cuentas que son de Estados Unidos y las que no lo son
    cuentas_eeuu = dataframe[dataframe['País'] == 'EstadosUnidos'].shape[0]
    cuentas_no_eeuu = dataframe[dataframe['País'] != 'EstadosUnidos'].shape[0]

    # Datos para el gráfico circular
    labels = ['Estados Unidos', 'No Estados Unidos']
    sizes = [cuentas_eeuu, cuentas_no_eeuu]
    colors = ['#ff9999','#66b3ff']
    explode = (0.1, 0)  # Solo "explode" el primer slice

    # Crea el gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
            shadow=True, startangle=140)
    plt.title('Porcentaje de cuentas de Instagram por país')
    plt.axis('equal')  # Para asegurar que el gráfico sea circular
    plt.show()

grafico_pais()
```

```python
# Inspecciona los valores únicos en la columna 'País' después de la normalización
print("Valores únicos después de la normalización:", dataframe['País'].unique())
```

```python
texto = """Ingrese la operación a realizar:

1. Indicar el influencer con más seguidores
2. Indicar los 3 países con más influencer
3. Indicar la cuenta de Marca con más seguidores: Nombre, cantidad de seguidores
4. Graficar seguidores por paises
5. Salir del menú
"""

opcion = int(input(texto))

if opcion == 1:
    mas_seguidores()
elif opcion == 2:
    paises()
elif opcion == 3:
    marca()
elif opcion ==4:
    grafi()
elif opcion == 5:
    print("Usted salió del menú")
else:
    print("Valor ingresado incorrecto")
```
